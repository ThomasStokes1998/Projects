```python
def elo(home, away, lam = 0.02):
    h = 1 - 1 / (1+np.exp(lam*(away-home)))
    a = - 1 / (1+np.exp(lam*(home-away)))
    return [h, a]
```

```python
events = wc2019.eventId.unique() # List of unique events
elo_dict = {} # creates a matrix of the elo rating for every competitor in every event
for wcaid in wc2019.personId.unique():
    elo_dict[wcaid] = [-1]*wc2019.eventId.nunique() # default is -1 so these elo ratings won't show up in tables
for e in range(len(events)): # using an index so I can access the list easier
    event = events[e]
    wc2019e = wc2019[wc2019.eventId == event].reset_index(drop="index")
    for rtype in wc2019e.roundTypeId.unique():
        wc2019er = wc2019e[wc2019e.roundTypeId == rtype].reset_index(drop="index")
        for i in range(len(wc2019er.pos)): # filters through all the competitors in order
            homewcaid = wc2019er.personId[i]
            if elo_dict[homewcaid][e] == -1:
                elo_dict[homewcaid][e] = 1000 # default elo is 1000
            for j in range(len(wc2019er.pos)):
                if i < j: # so we know that the 'home' competitor is ranked higher than the 'away' competitor
                    awaywcaid = wc2019er.personId[j]
                    if elo_dict[awaywcaid][e] == -1:
                        elo_dict[awaywcaid][e] = 1000
                    points = elo(elo_dict[homewcaid][e], elo_dict[awaywcaid][e]) # points to be added or subtracted
                    elo_dict[homewcaid][e] += 1600/len(wc2019er) * points[0] # points are adjusted for the size of the round
                    elo_dict[awaywcaid][e] -= 1600/len(wc2019er) * points[1]
elo_table = {}
elo_table["personId"] = wc2019.personId.unique()
for e in range(len(events)):
    event = events[e]
    elo_table[event] = []
    for wcaid in wc2019.personId.unique():
        score = elo_dict[wcaid][e]
        elo_table[event].append(score)
        
elo_df = pd.DataFrame(elo_table)
```

```python
def day(x):
    d = int(x/(24*3600))
    hr = x - d*24*3600
    h = int(hr/3600)
    minsec = hr - h*3600
    m = int(minsec/60)
    sec = minsec - m*60
    s = round(sec, 2)
    return str(d)+':'+str(h)+':'+str(m)+':'+str(s)
```

```python
time = []
for t in dg['Total_Time(s)']:
    time.append(day(t))
    
dg['Time(dd:hh:mm:ss)'] = time
dg
```

```python
def name(wcaid):
    row = persons[persons.id == wcaid]
    return row['name']
```

```python
ev_list = ['333', '333oh', '444', '555', 'minx', 'clock', '222', 'sq1', 'pyram', 'skewb']
wcaId = []
n = []
time = []
for x in ev_list:
    data = event(x).sort_values('total').reset_index(drop='index')
    wcaid = ''
    tim = 3600
    for i in range(len(data.eventId)):
        wid = data.personId[i]
        avg = [data.value1[i], data.value2[i], data.value3[i], data.value4[i], data.value5[i]]
        best = np.min(avg)
        worst = np.max(avg)
        decent = worst
        if best > 0 and best < tim:
            for t in avg:
                if best < t < worst and t < decent:
                    decent = t 
        if best == 0 and np.quantile(avg, 0.25) > 0:
            decent = np.quantile(avg, 0.5)
        if decent == 0:
            decent = 3600
        if decent < tim:
            wcaid = wid
            tim = decent
    wcaId.append(wcaid)
    n.append(name(wcaid))
    time.append(tim)
        
best_counting = pd.DataFrame({'eventId':ev_list, 'wcaId':wcaId, 'name':n, 'Counting_Time':time})
```
